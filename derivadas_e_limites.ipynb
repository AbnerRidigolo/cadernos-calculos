```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
```

```python
x = smp.Symbol('x')
```

```python
funcao = ((x**2 + x - 2) / (x - 1))
```

```python
funcao
```

```python
smp.limit(funcao, x, 1)
```

```python
smp.plot(funcao)
```

Estudos de limites da função logística

```python
x = smp.Symbol('x')
x
```

```python
funcao = 1 / (1+smp.exp(-x))
funcao
```

```python
#limite da função quando X tende a 0
smp.limit(funcao, x, 0)
```

```python
#Limite da função quando x tende a Infinito
smp.limit(funcao, x, smp.oo)
```

```python
#Limite da função quando x tende a Infinito negativo
smp.limit(funcao, x, -smp.oo)
```

```python
# obter a derivada
x = smp.Symbol('x')
f = x**2
smp.diff(f)
# graficos
funcao_original = x**2
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
# grafico funcao original
y = [funcao_original(i) for i in range(-100, 100)]
plt.plot(y)
# grafico funcao derivada
y_d = [funcao_derivada(i) for i in range(-100, 100)]
plt.plot(y_d)
```

```python
# obter a derivada
x = smp.Symbol('x')
f = (x**2 + 3) * (2*x + 1)
f_d = smp.diff(f)
smp.simplify(f_d)
# graficos
funcao_original = (x**2 + 3) * (2*x + 1)
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
# funcao original
y = [funcao_original(i) for i in range(-100, 100)]
plt.plot(y)
# funcao derivada
y_d = [funcao_derivada(i) for i in range(-100, 100)]
plt.plot(y_d)
```

```python
# obter a derivada
x = smp.Symbol('x')
f = (3*x**2 + 1) / (2*x - 1)
f_d = smp.diff(f)
smp.simplify(f_d)
# graficos
funcao_original = (3*x**2 + 1) / (2*x - 1)
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
# grafico funcao original
y = [funcao_original(i) for i in range(-10, 10)]
plt.plot(y)
# grafico funcao derivada
y_d = [funcao_derivada(i) for i in range(-10, 10)]
plt.plot(y_d)
```
