## 1 - Módulos e dados

- Módulos

```python
import numpy as np
import pandas as pd
```

- Dados

```python
dados = pd.read_csv('dados_1997_2011_paises_csv.csv',
                    sep = ';',
                    decimal = ',',
                    encoding = 'latin1')
```

## 2 - Análise por país

Obter: média, mediana, mínimo, máximo, quartis 1 e 3 e desvio padrão para as colunas “idh” e “competitividade_indice” de todos os países.

- IDH

```python
dados.groupby('pais') \
     .agg(min_idh = pd.NamedAgg('idh', 'min'),
          max_idh = pd.NamedAgg('idh', 'max'),
          media_idh = pd.NamedAgg('idh', 'mean'),
          mediana_idh = pd.NamedAgg('idh', 'median'),
          dp_idh = pd.NamedAgg('idh', 'std'))
```

```python
dados.groupby('pais')['idh'] \
     .apply(lambda x: x.quantile([0.25, 0.75])) \
     .to_frame() \
     .reset_index() \
     .rename(columns={'level_1': 'percentil'})
```

- Competitividade

```python
dados.groupby('pais') \
     .agg(min_compet = pd.NamedAgg('competitividade_indice', 'min'),
          max_compet = pd.NamedAgg('competitividade_indice', 'max'),
          media_compet = pd.NamedAgg('competitividade_indice', 'mean'),
          mediana_compet = pd.NamedAgg('competitividade_indice', 'median'),
          dp_compet = pd.NamedAgg('competitividade_indice', 'std'))
```

```python
dados.groupby('pais')['competitividade_indice'] \
     .apply(lambda x: x.quantile([0.25, 0.75])) \
     .to_frame() \
     .reset_index() \
     .rename(columns={'level_1': 'percentil'})
```
